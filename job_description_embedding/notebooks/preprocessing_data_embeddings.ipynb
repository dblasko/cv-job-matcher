```python
#just trying
import xml.etree.ElementTree as ET
import spacy

# Load the pre-trained English language model for Named Entity Recognition
nlp = spacy.load('en_core_web_sm')

def preprocess_xml_file(file_path):
    # Parse the XML file
    tree = ET.parse(file_path)
    root = tree.getroot()
    
    # Create a list to store the extracted features
    features = []
    
    # Iterate over each <job> element in the XML
    for job in root.iter('job'):
        # Extract relevant information from the XML
        title = job.find('title').text.strip()
        company = job.find('company').text.strip()
        body = job.find('body').text.strip()
        
        # Perform Named Entity Recognition on the job description
        doc = nlp(body)
        skills = []
        experience = []
        education = []
        
        # Extract entities and categorize them into skills, experience, and education
        for ent in doc.ents:
            if ent.label_ == 'SKILL':
                skills.append(ent.text)
            elif ent.label_ == 'EXPERIENCE':
                experience.append(ent.text)
            elif ent.label_ == 'EDUCATION':
                education.append(ent.text)
        
        # Create a dictionary to store the extracted features
        job_features = {
            'title': title,
            'company': company,
            'skills': skills,
            'experience': experience,
            'education': education
        }
        
        # Append the job features to the list
        features.append(job_features)
    
    return features

#XML file
file_path = 'job_sample.xml'

preprocessed_data = preprocess_xml_file(file_path)
for job in preprocessed_data:
    print(job)
```
